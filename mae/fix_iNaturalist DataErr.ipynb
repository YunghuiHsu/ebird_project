```python
import json
import os
from pathlib import Path
import pandas as pd
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from util.crop import RandomResizedCrop

data_path = Path('../ebird/iNaturalist_2021')
```

```python
transform_train = transforms.Compose([
        RandomResizedCrop(224, interpolation=3),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])
transform_val = transforms.Compose([
        transforms.Resize(256, interpolation=3),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])
dataset_train = datasets.ImageFolder(data_path/'train', transform=transform_train)
dataset_val = datasets.ImageFolder(data_path/'val', transform=transform_val)
```

```python
tmp = dataset_train.class_to_idx
train_cls_idx = dict(zip(tmp.values(), tmp.keys()))

tmp = dataset_val.class_to_idx
val_cls_idx  = dict(zip(tmp.values(), tmp.keys()))

print('Number of dir in train :',   len(train_cls_idx))
print('Number of dir in val :',  len(val_cls_idx))
```

```python
for i, idx in enumerate(train_cls_idx.keys()):
    cls_train, cls_val =  train_cls_idx[idx], val_cls_idx[idx]
    
    print(idx, cls_train )
    print(idx,  cls_val)
```

```python
train_subdir = {}
for idx, entry in enumerate(os.scandir(data_path/'train')):
    train_subdir[idx] = entry.name
    print(idx, entry.name, end='\r')
```
